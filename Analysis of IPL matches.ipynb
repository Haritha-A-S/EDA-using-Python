# Analysis of IPL Matches

```python
# loading required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading IPL match dataset
df = pd.read_csv("matches.csv")
# having glance at first five rows
df.head()
```

```python
# having glance on data 
df.info()
```

```python
#Lookin at the number of rows and columns in the dataset
df.shape
```

```python
#Getting the frequency of most man of the match awards
df['player_of_match'].value_counts()
```

```python
#Getting the top 10 players with most man of the match awards
df['player_of_match'].value_counts()[0:10]
```

```python
#Getting the top 5 players with most man of the match awards
df['player_of_match'].value_counts()[0:5]
```

```python
#making a bar-plot for the top 5 players with most man of the match awards
plt.figure(figsize=(10,5))
plt.bar(list(df['player_of_match'].value_counts()[0:5].keys()),list(df['player_of_match'].value_counts()[0:5]),color='maroon')
plt.show()
```

```python
#Getting the frequency of result column
df['result'].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
df['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=df[df['win_by_runs']!=0]
batting_first.head()
```

```python
#Making a histogram 
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'],color='maroon')
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 5 teams with most wins after batting first
plt.figure(figsize=(15,4))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color=["blue","yellow","orange","green","purple"])
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extracting those records where a team has won after bowling_first
bowling_first=df[df['win_by_wickets']!=0]
bowling_first.head()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(5,5))
plt.hist(bowling_first['win_by_wickets'],bins=30)
plt.show()
```

```python
#Finding out the frequency of number of wins w.r.t each time after bowling_first
bowling_first['winner'].value_counts()
```

```python
#Making a bar plot for top-5 teams with most wins after batting second
plt.figure(figsize=(15,4))
plt.bar(list(bowling_first['winner'].value_counts()[0:5].keys()),list(bowling_first['winner'].value_counts()[0:5]),color=["green","blue","purple","yellow","pink"])
plt.show()
```

```python
#Making a pie chart for distribution of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(list(bowling_first['winner'].value_counts()),labels=list(bowling_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
