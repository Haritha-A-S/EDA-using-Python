# EDA FIFA analysis

```python
#loading libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading and opening the file

fifa = pd.read_csv("players_20.csv")
fifa.head()
```

```python
fifa.info()
```

```python
fifa.describe()
```

```python
for col in fifa.columns:
    print(col)
```

```python
# finds records(rows, columns)
fifa.shape
```

```python
# Checking the nationality of players

fifa['nationality'].value_counts()
```

```python
# Checking the top 10 countries for nationality values

fifa['nationality'].value_counts()[0:10]
```

```python
# Checking the top 10 countries for nationality names i.e keys

fifa['nationality'].value_counts()[0:10].keys()
```

```python

plt.figure(figsize = (15,7))
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()),list(fifa['nationality'].value_counts()[0:10]), color= "purple")
plt.show()
```

Inference :
From bar graph it is clear that most of the players belong to England.

```python
# checking player vs salary by creating player_salary data frame

player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
player_salary = player_salary.sort_values(['wage_eur'],ascending=False)
player_salary.head()
```

```python

plt.figure(figsize = (17,7))
plt.bar(list(player_salary['short_name'][0:10]),list(player_salary['wage_eur'][0:10]), color= "purple")
plt.show()
```

Inference : It is observed that L.Messi has highest wage

```python
# Finding which player as best shooting skills in decending order

player_shooting = fifa[['nationality','short_name','shooting','club']]
player_shooting.sort_values(['shooting'],ascending=False).head()
```

```python
# Finding which player as best defending skills in decending order

player_defending = fifa[['nationality','short_name','defending','club']]
player_defending.sort_values(['defending'],ascending=False).head()
```
