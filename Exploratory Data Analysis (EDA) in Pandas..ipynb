# Exploratory Data Analysis (EDA) in Pandas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Population dataset.xlsx')
df
```

```python
pd.set_option('display.float_format',lambda x: '%.2f' % x)
```

```python
df.info()
```

```python
# some missing values are observed as their column count is not same.
```

```python
df.describe()
# this gives an over view
```

```python
df.isnull().sum()

#gives count of null values wrt columns
```

```python
df.nunique()
```

```python
df.sort_values(by= "World Population Percentage", ascending=False).head(10)
# sorting using 2022 population in asending order by default. To remove that we use ascending = False.Then it will give descending order
#sorting using world population percentage
```

```python
# checking correlation
df.corr()
```

```python
# using Heat map to visualize
# If annot=True then map displays its values also ,otherwise shows only colours
# plt.rcParams['figure.figsize']=(30,20)     this resizes the map

sns.heatmap(df.corr(),annot = True)
plt.rcParams['figure.figsize']=(10,7)
plt.show()
```
